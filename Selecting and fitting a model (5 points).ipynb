#### 2. Bias-variance decomposition<br/>
a. Generate a graph of typical (squared) bias, variance, training error, test error, and Bayes (irreducible) error curves, on a single plot, as we go from less flexible statistical learning methods towards more flexible approaches. The x-axis should represent the amount of flexibility in the method, and the y-axis should represent the values for each curve. There should be five curves. Make sure to label each one.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 20, 0.1)
y1 = 60 / 1.5 ** (x - 0.5) + 55
y2 = 0.03 * x ** 3 + 10
y3 = 65 / 2 ** (x - 0.5) + 35
y4 = 50 / 1.5 ** x + 0.03 * x ** 3 + 40
y5 = 0 * x + 55

plt.plot(x, y1, label='Typical(squared) Bias', color='blue')
plt.plot(x, y2, label='Variance', color='yellow')
plt.plot(x, y3, label='Training Error', color='green')
plt.plot(x, y4, label='Test Error', color='black')
plt.plot(x, y5, label='Bayes Error', color='orange')
plt.xlabel('Flexibility')
plt.ylabel('Mean Squared Error')
plt.legend()
plt.show()
```

b. Explain why each of the five curves has the shape displayed in part (a).<br/>
Bias will decrease if we use a more flexible model. As we can see from the plot, the squared bias will decrease as flexibility increasing. And variance will become bigger in a more flexible model. This is because the flexible model follows closely to the data, and a minor variation would cause the estimation to change significantly. As the model becomes more flexible, the model will try to fit more data and the training error will decrease. The test error is dependent on two different forces: one is variance and the other is bias. When the model is relatively inflexible, the bias is large and the variance is small. And the bias will decrease while variance will increase when the model becomes more flexible. Bayes error will remain the same according to the flexibility, because any model cannot completely describe the real data.

```python
np.random.seed(50)
```

```python
x1 = np.random.uniform(-1, 1, 200)
x2 = np.random.uniform(-1, 1, 200)
```

```python
u = np.random.normal(0, 0.5, 200)
y = x1 + x1 ** 2 + x2 + x2 ** 2 + u
```

```python
pb_suc = np.exp(y) / (1 + np.exp(y))
```

```python
suc = pb_suc > 0.5
fail = pb_suc <= 0.5
plt.scatter(x1[suc], x2[suc], label = 'success')
plt.scatter(x1[fail], x2[fail], label = 'failure')
plt.legend()

X, Y = np.meshgrid(np.linspace(-1, 1, 200), np.linspace(-1, 1, 200))
yy = X + X * X + Y + Y * Y
prob = np.exp(yy) / (1 + np.exp(yy))
plt.contour(X, Y, (prob>0.5).astype(np.int), linestyles='dashed')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Simulation and Bayes Decision Boundary')
plt.show()
```
